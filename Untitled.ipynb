```python
import pandas
import requests
import io
```

```python
base_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/'
conf_file = 'time_series_19-covid-Confirmed.csv'
deaths_file = 'time_series_19-covid-Deaths.csv'
recov_file = 'time_series_19-covid-Recovered.csv'
```

```python
def get_table_from_url(url):
    csv_raw_text = requests.get(url).text
    csv_buffer = io.StringIO(csv_raw_text)
    
    return pandas.read_csv(csv_buffer)
```

```python
conf = get_table_from_url(base_url + conf_file)
conf.tail()
```

```python
deaths = get_table_from_url(base_url + deaths_file)
deaths.tail()
```

```python
recov = get_table_from_url(base_url + recov_file)
recov.tail()
```
